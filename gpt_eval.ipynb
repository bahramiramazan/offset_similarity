```python

dataname='semeval_2012_hard'
file='essential_files/gpt/answers'+dataname+'_40-mini.json'
#file='../essential_files/answers'+dataname+'_gpt-5-nano.json'
import json
data= json.load(open(file))
#h_data={'all_answers':all_answers,'solution_set':solution_set,'all_questions':all_questions,'batch':batch}
all_answers=data['all_answers']
solution_set=data['solution_set']
all_questions=data['all_questions']
```

```python
from sklearn.metrics import f1_score

all_questions
eval_d='science'#metaphor #science
c=0
nc=0
Y=[]
PREDICTIONS=[]
for ki,k in enumerate(all_answers):
    a=all_answers[k]
    k=k.split('NO')[-1]
    
    
    eval_d_=all_questions[int(ki)]['questions']#['eval_d']

    if eval_d_!=eval_d and dataname=='scan' :
        continue
    if k not in solution_set.keys():
        continue
 
    sol=solution_set[k][-1]
    if a==sol:
        c=c+1
    else:
        q=all_questions[int(k)]
        #eval_d_=all_questions[int(ki)]['questions']#['eval_d']
        print(sol,a)
        nc=nc+1
    if a==None:
        continue
    Y.append(sol)
    print('a',a)
    PREDICTIONS.append(a)
    
        
print('al',c,nc,c/(c+nc))
print(ki)
f1_macro=f1_score(Y, PREDICTIONS, average='macro')
f1_micor=f1_score(Y, PREDICTIONS, average='micro')
print('f1_macro',f1_macro)
print('f1_micor',f1_micor)
```
